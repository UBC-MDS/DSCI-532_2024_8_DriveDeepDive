```python
import numpy as np
import pandas as pd
from uszipcode import SearchEngine
import re
```

```python
df = pd.read_csv('../data/raw/used_car_sales.csv')
```

```python
df_us = df[df['zipcode'].astype(str).str.len() == 5]
df_us
```

```python
def modify_zipcode(zipcode):
    pattern = r'\d{3}\*\*'
    if re.match(pattern, zipcode):
        return re.sub(r'\*\*', '01', zipcode)
    else:
        return zipcode

# Applying the modification function to the DataFrame
df_us['modified_zipcode'] = df_us['zipcode'].apply(modify_zipcode)
df_us
```

```python
search = SearchEngine()
def map_zipcode_to_state(zipcode):
    result = search.by_zipcode(zipcode)
    if result and result.state:
        return result.state
    else:
        return "Invalid or Unknown Zipcode"

df_us['state'] = df_us['modified_zipcode'].apply(map_zipcode_to_state)
```

```python
df_processed = df_us[df_us['state'] != 'Invalid or Unknown Zipcode']
df_processed
```
